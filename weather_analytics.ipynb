# 🌤️ Weather Agent - Advanced Analytics Notebook

This notebook connects to your live **Weather Platform API** on Render to perform deep analysis and visualization of search trends and agricultural metrics.

### 🔗 API Configuration
Replace the `API_URL` below with your Render service URL if it changes.

```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Your live Render API
API_URL = "https://weather-1-5aek.onrender.com/api/analytics"

print(f"Pulling live data from: {API_URL}...")
```

## 📥 Data Fetching

```python
response = requests.get(API_URL)
data = response.json()

if data['success']:
    print("✅ Data retrieved successfully!")
    print(f"Total Searches: {data['totalSearches']}")
    print(f"Active Geo-Nodes: {data['geoNodes']}")
    print(f"System Uptime: {data['uptime']} seconds")
else:
    print("❌ Failed to fetch data.")
```

## 📊 Search Distribution (Top Cities)

```python
df_cities = pd.DataFrame(data['topCities'])

plt.figure(figsize=(10, 6))
sns.set_theme(style="darkgrid")
ax = sns.barplot(x="searchQuery", y="count", data=df_cities, palette="viridis")
plt.title("Top Searched Regions", fontsize=16)
plt.xlabel("City/Query", fontsize=12)
plt.ylabel("Volume", fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

## 🌱 Agricultural Insights
Aggregated soil metrics from global telemetry.

```python
metrics = {
    "Avg Soil Temp (°C)": float(data['avgSoilTemp']),
    "Avg Soil Moisture (m³/m³)": float(data['avgSoilMoisture'])
}

plt.figure(figsize=(8, 5))
plt.bar(metrics.keys(), metrics.values(), color=['#f87171', '#60a5fa'])
plt.title("Global Soil Telemetry Average", fontsize=14)
plt.ylabel("Metric Value")
plt.show()
```
